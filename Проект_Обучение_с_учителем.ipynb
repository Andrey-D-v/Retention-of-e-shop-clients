## Пайплайны

**Разделим данные на обучающую и тестовую выборки. Учтем при этом необходимость равномерного распределения между выборками значений целевого признака.**

```python
X = df.drop('покупательская_активность', axis=1)
y = df['покупательская_активность']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
```

**В целевом признаке "покупательская активность" закодируем "Прежний уровень" как 0, "Снизилась" - как 1.**

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

**Сформируем пайплайн, в рамках которого категориальные столбцы будут обрабатываться OneHotEncoder и OrdinalEncoder, а столбцы с количественными признаками - масштабироваться  с помощью StandardScaler и MinMaxScaler. Переберем 4 варианта моделей (DecisionTreeClassifier, KNN, LogisticRegression и SVM) с различными гиперпараметрами, типичными для этих типов моделей. Лучшую модель определим при помощи рассчитанной на кросс-валидации метрики ROC-AUC. Данная метрика нам хорошо подходит, так как она одновременно оценивает модель по тому, насколько верно или ошибочно она предсказывает класс 1 ("покупательская активность снизилась") - True Positive Rate и False Positive Rate. Также ROC-AUC не чувствительна к дисбалансу классов, что поможет нам избежать негативного влияния того факта, что у нас значений класса 1 на 20 % меньше, чем класса 0.**

```python
ohe_columns = ['разрешить_сообщать', 'популярная_категория']
ord_columns = ['тип_сервиса']
num_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность', 'акционные_покупки',
              'средний_просмотр_категорий_за_визит', 'неоплаченные_продукты_штук_квартал', 'ошибка_сервиса',
               'страниц_за_визит', 'предыдущий_месяц_выручка', 'препредыдущий_месяц', 'текущий_месяц_выручка',
               'предыдущий_месяц_минуты', 'текущий_месяц_минуты']
```

```python
ohe_pipe = Pipeline(steps=[
    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False))])

ord_pipe = Pipeline(steps=[
    ('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ord', OrdinalEncoder(categories=[['стандарт', 'премиум']], handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))])

data_preprocessor = ColumnTransformer(transformers=[
    ('ohe', ohe_pipe, ohe_columns),
    ('ord', ord_pipe, ord_columns),
    ('num', StandardScaler(), num_columns)
], remainder='passthrough')

pipe_final = Pipeline(steps=[
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=0))
])

param_grid = [{'models' : [DecisionTreeClassifier(random_state=0)],
               'models__max_depth' : range(2,10),
               'models__max_features' : range(2,10),
               'preprocessor__num' : [StandardScaler(), RobustScaler()]},
              {'models' : [LogisticRegression(random_state=0, solver='saga')],
               'models__penalty' : ['l1', 'l2', 'elasticnet'],
               'models__C' : [0.1, 0.5, 1, 5],
               'preprocessor__num' : [StandardScaler(), RobustScaler()]},
              {'models' : [SVC(random_state=0, probability=True)],
               'models__C' : [0.1, 0.5, 1, 5],
               'models__kernel' : ['rbf', 'poly', 'linear'],
               'preprocessor__num' : [StandardScaler(), RobustScaler()]}]

grid_search = GridSearchCV(pipe_final, param_grid=param_grid, scoring='roc_auc', cv=3,
                          n_jobs=-1, refit=True)

grid_search.fit(X_train, y_train)
y_test_pred = grid_search.predict(X_test)
```

**Выведем параметры лучшей модели, а также значения метрики ROC-AUC для обучающей и тестовой выборки.**

```python
grid_search.best_params_
```

```python
X_tr = data_preprocessor.fit_transform(X_train)
X_tt = data_preprocessor.transform(X_test)

probabilities_tr = grid_search.best_estimator_._final_estimator.predict_proba(X_tr)
probabilities_tt = grid_search.best_estimator_._final_estimator.predict_proba(X_tt)
probabilities_train = probabilities_tr[:, 1]
probabilities_test = probabilities_tt[:, 1]
print('Площадь ROC-кривой обучающей выборки:', roc_auc_score(y_train, probabilities_train))
print('Площадь ROC-кривой тестовой выборки:', roc_auc_score(y_test, probabilities_test))
```

**В качестве лучшей модели определена модель опорных векторов, причем с дефолтными параметрами (RBF-алгоритм и значение регуляризации - 1). Масштабирование данных выполняется с помощью StandardScaler. Значение метрики на обучающей выборке - 0.96, тестовой выборке - 0.94.**

```python
pd.DataFrame(grid_search.cv_results_).sort_values(by='rank_test_score').head()
```

**Судя по матрице ошибок, модель очень хорошо определяет класс 0.**

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, y_test_pred,
                                        labels=grid_search.best_estimator_._final_estimator.classes_)
                                        ).plot();
```

**ВЫВОД: Наилучшей по метрике ROC-AUC (0.96 на обучающей и 0.94 на тестовой выборке) оказалась модель SVM с дефолтными параметрами, наилучшим инструментом масштабирования - StandardScaler.**

## Анализ важности признаков

**Чтобы заглянуть внутрь модели и определить, какие признаки она считает значимыми, построим график распределения значений Шепли.**

В свою очередь, признак "акционные покупки" влияет на результат в противоположную сторону: чем выше доля акционных покупок у клиента, тем выше вероятность его отнесения к классу 1. Выявленные значимые признаки явно указывают на то, что сохранению покупательской активности клиентов содействуют их более активные коммуникации с сайтом (в плане времени, количества просмотренных страниц и категорий). При этом тенденцию к снижению покупательской активности демонстрируют клиенты с более высокой долей акционных покупок.**

## Сегментация покупателей

**Для сегментации клиентов объединим датасет обучающей выборки с данными о прибыльности клиентов и вероятности попадания их в класс 1 и оставим в таблице только клиентов, чья покупательская активность снизилась.**

```python
X_train_c = X_train.copy()
X_train_c['probability_1'] = probabilities_train
X_train_c = X_train_c.merge(profit, how='left', left_index=True, right_index=True)
X_train_c_2 = X_train_c[X_train_c['probability_1'] >= 0.5]
X_train_c_2.head()
```

**График зависимости между вероятностью отнесения к классу 1 и прибыльностью клиентов указывает на наличие в выборке обширной группы покупателей разной прибыльности с крайне высокой вероятностью снижения активности (выше 0.9).**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=X_train_c_2, x='probability_1', y='прибыль');
```

**Выделим в этой группе сегмент покупателей с высокой прибыльностью (т.е. выше медианного значения прибыли в целом по клиентской базе - 4 ед.). Таких клиентов оказалось 136.**

```python
segment = X_train_c_2[(X_train_c_2['probability_1'] > 0.9) & (X_train_c_2['прибыль'] > 4)]
print('Количество клиентов в сегменте - ', len(segment))
```

**Сравним медианные значения признаков в данном сегменте с медианными значениями признаков в целом по клиентской базе.**

```python
comparison = pd.concat([X_train_c.median(), segment.median()], axis=1)
comparison.columns = ['все клиенты', 'клиенты сегмента']
comparison
```

**Также посмотрим, есть ли какие-либо особенности в категориальных признаках в выделенном сегменте.**

```python
cat_features = ['тип_сервиса', 'популярная_категория', 'разрешить_сообщать']
for feature in cat_features:
    print(feature)
    print(round(segment[feature].value_counts(normalize=True), 2))
    print()
```

Заметно также, что в отношении этих клиентов в последние полгода маркетинговая активность была ниже средней. При этом в сегменте на 10 п.п. выше показатель доли акционных покупок.**

**В этой связи сформулируем следующие предложения по улучшению покупательской активности выделенного клиентского сегмента: 1.Расширить / повысить интенсивность акционных программ. 2.Нарастить маркетинговую активность в отношении данного сегмента покупателей. Эти меры призваны повысить интерес клиентов к сайту, предлагаемым товарам, из которых клиенты с учетом акций выберут для себя то, что им покажется интересным. Фактор неприязни к рекламе в сегменте отсутствует: не возражают против рассылок 3/4 клиентов - как в среднем по выборке. В плане товарных категорий стоит продвигать только 3 наиболее популярных среди всех клиентов категории - товары для детей, домашний текстиль и косметика и аксессуары; остальные же категории лучше не задействовать вовсе.**
